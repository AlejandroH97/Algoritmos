```python
import pylab as pl
import random
```

```python
def random_perm(n):
    v = [i for i in range(n)]
    for i in range(len(v)-1):
        j = random.randint(i, len(v)-1)
        v[i], v[j] = v[j], v[i]
    return v
```

```python
def Bubble_sort(N,lista):
    ins = 0
    comp = 0
    swaps = 0
    for i in range(N-1):
        ins += 1
        for x in range(N-i-1):
            if lista[x] > lista[x+1]:
                comp+= 1
                lista[x],lista[x+1] = lista[x+1],lista[x]
                ins += 2
                swaps += 1
            ins+= 2
    return ins, comp, swaps
```

```python
max = 9
for n in range(max+1):

    instruccionespromedio = []
    comparasionespromedio = []
    swapspromedio = []
    runs = 25000

    for n in range(n, (n+1)):
        for _ in range(runs):
            p = random_perm(n)
            psorted = Bubble_sort(len(p),p)
            instruccionespromedio.append(psorted[0])
            comparasionespromedio.append(psorted[1])
            swapspromedio.append(psorted[2])

    pl.hist(instruccionespromedio, bins='auto')
    pl.title("Histograma de pasos con todas las permutaciones de N = %s" % (n))
    pl.show()

    pl.hist(comparasionespromedio, bins='auto')
    pl.title("Histograma de comparaciones con todas las permutaciones de N = %s" % (n))
    pl.show()

    pl.hist(swapspromedio, bins='auto')
    pl.title("Histograma de swaps con todas las permutaciones de N = %s" % (n))
    pl.show()
```
