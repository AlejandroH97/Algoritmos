```python
import pylab as pl
import random
```

```python
def random_perm(n):
    v = [i for i in range(n)]
    for i in range(len(v)-1):
        j = random.randint(i, len(v)-1)
        v[i], v[j] = v[j], v[i]
    return v
```

```python
def insertion_sort(lista):

    ins = 0
    comp = 0
    swaps = 0
    k = 1
    i = 1
    while (i < len(lista)):
        key = lista[k]
        ins += 3
        comp += 1
        while (k > 0) and (key < lista[k-1]):
            comp+= 1
            lista[k], lista[k-1] = lista[k-1], lista[k]
            swaps += 1
            k=k-1
            ins += 3
        i = i+1
        k = i
        ins += 1
    return ins, comp, swaps
```

```python
max = 10
for n in range(max):

    instruccionespromedio = []
    comparacionespromedio = []
    swapspromedio = []
    runs = 25000

    for n in range(n, n+1):

        for _ in range(runs):
            p = random_perm(n)
            psorted = insertion_sort(p)
            instruccionespromedio.append(psorted[0])
            comparacionespromedio.append(psorted[1])
            swapspromedio.append(psorted[2])

    pl.hist(instruccionespromedio, bins='auto')
    pl.title("Histograma de pasos con todas las permutaciones de N = %s" % (n))
    pl.show()

    pl.hist(comparacionespromedio, bins='auto')
    pl.title("Histograma de comparaciones con todas las permutaciones de N = %s" % (n))
    pl.show()

    pl.hist(swapspromedio, bins='auto')
    pl.title("Histograma de swaps con todas las permutaciones de N = %s" % (n))
    pl.show()
```
